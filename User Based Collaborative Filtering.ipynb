```python
movieratings = ratings.pivot_table(index = 'user_id', columns = ['title'], values = 'rating')
```

```python
# corr with, 2 dataframes and find correlation

movieratings.corrwith(starwars)
```

```python
similarmovies = movieratings.corrwith(starwars)
similarmovies  = similarmovies.dropna()
similarmovies
```

```python
similarmovies  = similarmovies.sort_values(ascending = False)
similarmovies

#title is index
```

```python
similarmovies.index
```

```python
moviestats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
moviestats.head()
```

# Some movies have been rated by only a few people so let's filter the result for movies which have greater than 100 reviews

```python
popularmovies = moviestats[moviestats['rating','size'] >= 100]
```

```python
popularmovies.sort_values(('rating','mean'), ascending = False)
```

```python
len(similarmovies),len(popularmovies)
```

```python
#joining index, similar titles, left join, similar is the corr

df = popularmovies.join(pd.DataFrame(similarmovies, columns = ['similar']))
```

```python
df.head()
```

```python
len(df)
```

```python
df.sort_values('similar', ascending = False)
```
